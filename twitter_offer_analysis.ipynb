```python
df_twitter_matched = pd.read_csv('./1006/twitter_matched.csv', engine = 'python')
```

```python
df_twitter_matched
```

```python
df_new = pd.DataFrame(columns=['domain', 'category', 'payout'])
```

```python
ddd = list(set(df_twitter_matched.domain.to_list()))
```

计算平均payout时只考虑美元（$开头）按比例分成、卢比、欧元之类的不考虑，然后对多个offer的payout取平均值

```python
for i in tqdm(ddd):
    category = df_twitter_matched[df_twitter_matched['domain'] == i].iloc[0,3]
    pay_sum = 0
    count = 0
    df_temp = df_twitter_matched[df_twitter_matched['domain'] == i]
    for j in range(len(df_temp)):
        try:
            if df_temp.iloc[j, -1].startswith("$"):
                money = df_temp.iloc[j, -1].split('/')[0].replace(' ','').replace('$', '')
                pay_sum += float(money)
                count += 1
        except:
            pass
    payout = pay_sum/count if count != 0 else 0
    items = {
        "domain": i,
        "category": category,
        "payout": payout
    }
    df_new.loc[len(df_new)] = items
```

```python
df_twitter_allurl = pd.read_csv('./files/twitter_all_urls.csv', engine = 'python')
df_twitter_allurl = df_twitter_allurl.reindex(columns=df_twitter_allurl.columns.to_list() + ['landing_page', 'landing_page_domain', 'isAffiliate', 'offer_title', 'offer_category', 'offer_payout'])
df_twitter_allurl.landing_page = None
df_twitter_allurl.landing_page_domain = None
df_twitter_allurl.isAffiliate = 0
df_twitter_allurl.offer_title = None
df_twitter_allurl.offer_category = None
df_twitter_allurl.offer_payout = None
df_twitter_allurl
```

```python
df_twitter_0915
```

```python
from tqdm import tqdm

for i in tqdm(range(len(df_twitter_allurl))):
    url = df_twitter_allurl.loc[i, 'url']
    if url in affiliate_url_list:   # 是affiliate URL
        df_twitter_allurl.loc[i, 'isAffiliate'] = 1
        df_twitter_allurl.loc[i, 'landing_page'] = df_twitter_0915[df_twitter_0915['url'] == url].iloc[0, 1]
        df_twitter_allurl.loc[i, 'landing_page_domain'] = df_twitter_0915[df_twitter_0915['url'] == url].iloc[0, -8]
        if df_twitter_allurl.loc[i, 'url'] in twitter_url_list: # 是affiliate URL且匹配成功
            sub_domain = df_twitter_0915[df_twitter_0915['url'] == url].iloc[0, -2]
            df_twitter_allurl.loc[i, 'offer_category'] = df_new[df_new['domain'] == sub_domain].iloc[0,1]
            df_twitter_allurl.loc[i, 'offer_payout'] = df_new[df_new['domain'] == sub_domain].iloc[0,2]
```
