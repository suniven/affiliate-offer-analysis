```python
ddd = list(set(df_youtube_matched.domain.to_list()))
```

计算平均payout时只考虑美元（$开头）按比例分成、卢比、欧元之类的不考虑，然后对多个offer的payout取平均值

```python
for i in tqdm(ddd):
    category = df_youtube_matched[df_youtube_matched['domain'] == i].iloc[0,3]
    pay_sum = 0
    count = 0
    df_temp = df_youtube_matched[df_youtube_matched['domain'] == i]
    for j in range(len(df_temp)):
        try:
            if df_temp.iloc[j, -1].startswith("$"):
                money = df_temp.iloc[j, -1].split('/')[0].replace(' ','').replace('$', '')
                pay_sum += float(money)
                count += 1
        except:
            pass
    payout = pay_sum/count if count != 0 else 0
    items = {
        "domain": i,
        "category": category,
        "payout": payout
    }
    df_new.loc[len(df_new)] = items
```

```python
df_youtube_allurl = pd.read_csv('./1006/youtube_all_url.csv', engine = 'python')
df_youtube_allurl = df_youtube_allurl.reindex(columns=df_youtube_allurl.columns.to_list() + ['landing_page', 'landing_page_domain', 'isAffiliate', 'offer_title', 'offer_category', 'offer_payout'])
df_youtube_allurl.landing_page = None
df_youtube_allurl.landing_page_domain = None
df_youtube_allurl.isAffiliate = 0
df_youtube_allurl.offer_title = None
df_youtube_allurl.offer_category = None
df_youtube_allurl.offer_payout = None
df_youtube_allurl
```

```python
from tqdm import tqdm

for i in tqdm(range(len(df_youtube_allurl))):
    url = df_youtube_allurl.loc[i, 'url']
    if url in affiliate_url_list:   # 是affiliate URL
        df_youtube_allurl.loc[i, 'isAffiliate'] = 1
        df_youtube_allurl.loc[i, 'landing_page'] = df_youtube_0915[df_youtube_0915['url'] == url].iloc[0, 1]
        df_youtube_allurl.loc[i, 'landing_page_domain'] = df_youtube_0915[df_youtube_0915['url'] == url].iloc[0, -4]
        if df_youtube_allurl.loc[i, 'url'] in youtube_url_list: # 是affiliate URL且匹配成功
            sub_domain = df_youtube_0915[df_youtube_0915['url'] == url].iloc[0, -3]
            df_youtube_allurl.loc[i, 'offer_category'] = df_new[df_new['domain'] == sub_domain].iloc[0,1]
            df_youtube_allurl.loc[i, 'offer_payout'] = df_new[df_new['domain'] == sub_domain].iloc[0,2]
```
